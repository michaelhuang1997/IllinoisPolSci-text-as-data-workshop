```python
import xml.etree.ElementTree as ET
```

```python

def extract_roderigos_speeches(file_path):
    # Load the XML document
    tree = ET.parse(file_path)
    root = tree.getroot()

    # Namespace dictionary to help navigate the XML file
    ns = {'tei': 'http://www.tei-c.org/ns/1.0'}

    # Extract Roderigo's speeches
    roderigos_speeches = []

    # Navigate to speeches where the speaker is Roderigo
    for sp in root.findall('.//tei:sp[@who="Roderigo"]', ns):
        speech = []
        for l in sp.findall('tei:l', ns):
            # Join words to reconstruct the lines of the speech
            line = ' '.join([w.text for w in l.findall('tei:w', ns)])
            speech.append(line)
        # Add the reconstructed speech to our list
        roderigos_speeches.append('\n'.join(speech))
    
    return roderigos_speeches

# Use the function
file_path = '/Users/michaelhuang/Dropbox/University of Chicago/2022 Winter/DIGS 30031/Texts/oth.xml'
speeches = extract_roderigos_speeches(file_path)
for idx, speech in enumerate(speeches, 1):
    print(f"Speech {idx}:\n{speech}\n")


```
